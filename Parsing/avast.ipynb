Напишем скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса [Avast Hack Ckeck](https://www.avast.com/hackcheck/).
Список email-ов задана переменной в начале кода:  
`EMAIL = [xxx@x.ru, yyy@y.com]`

В итоге будет формирован датафрейм со столбцами: <дата утечки> - <название> - <описание утечки> - <почта>.

Комментарии буду писать либо в самом коде, либо в markdown отдельной строкой. 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# можно вбить сюда свой email
EMAIL = ['xxx@x.ru', 'yyy@y.com']
```

```python
# смотрим структуру ответа на запрос на примере xxx@x.ru
response = requests.post('https://digibody.avast.com/v1/web/leaks', json={ 'email': 'xxx@x.ru' })
# response.json()
```

```python
#создаем датафрейм, куда распаковываем словарь value
leaks = pd.DataFrame(response.json()['value'])
leaks
```

```python
#создаем датафрейм со словарём, который содержится в столбце leak_info датафрейма leaks (словарь)
leaks2 = leaks['leak_info'].apply(pd.Series)
leaks2
```

```python
#для удобства чтения меняем дату с unix на читаемый вид
leaks2['date'] = pd.to_datetime(leaks2['date'], unit='ms')
```

Создаю резюмирующий датафрейм res, куда помещяю уже проведённые выше действия (функционально они не нужны) и добавляю циклы:
 - первый цикл for el in EMAIL выполняет все предыдущие действия для каждого электронного адреса;
 - второй цикл for index, row in leak3.iterrows() добавляет результат выполнения предыдущего цикла в датафрейм res, который и выводим на экран. 

```python
res = pd.DataFrame()
for el in EMAIL:
    req = requests.post('https://digibody.avast.com/v1/web/leaks', json={'email': el })
    leak = pd.DataFrame(req.json()['value'])
    leak2 = leak['leak_info'].apply(pd.Series)
    leak2['date'] = pd.to_datetime(leak2['date'], unit='ms')
    leak2['email'] = el
    leak3 = leak2.loc[:, ['date', 'title', 'description', 'email']]
    for index, row in leak3.iterrows():
        df = pd.DataFrame([row])
        res = pd.concat([res, df])
res.index = range(len(res.index))
res
```
